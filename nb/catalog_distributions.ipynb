```python
from astropy.io import fits
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_histograms(filename):
    hdus = fits.open(filename)
    ra   = hdus[1].data['ra']
    dec  = hdus[1].data['dec']
    z    = hdus[1].data['z']
    w    = hdus[1].data['weight']
    
    plt.figure(1, figsize=(14, 14)) 
    plt.subplot(221)
    plt.hist(ra, bins=100)
    plt.xlim(0, 360)
    plt.xlabel("RA [deg]")
    plt.subplot(222)
    plt.hist(dec, bins=100)
    plt.xlabel("Dec [deg]")
    plt.subplot(223)
    plt.hist(z, bins=100)
    plt.xlabel("z")
    plt.show()
```

```python
plot_histograms("../samples/mocks_DESI_0.fits")
```

```python
plot_histograms("../samples/randoms_DESI_0.fits")
```
